## Types of Numbers

Python has the following types of numbers:
* int
* float
* complex (see docs if interested, not gonna cover)

[Python docs on numbers](https://docs.python.org/3/library/stdtypes.html?highlight=int%20float%20complex#numeric-types-int-float-complex)

### Demo: Number types

````python
# define an int
thisVarIsAnInt = 5
print(f"thisVarIsAnInt is: {thisVarIsAnInt}")

# define a float
floatNum = 5.0
print(f"floatNum is: {floatNum}")

# you can do math regardless of type
print(f"adding float and int yields: {thisVarIsAnInt + floatNum}")

# math with two ints returns a float.  
# This is different than how it worked in python2
print(f"14 / 5: {14 / 5}")

# two ints return int
print(f"5*5 : {5*5}")

# by including a float in the operation will always return a float
print(f"5*5.0: {5*5.0}")

# convert int to float
varInt = 10
varFloat = float(10)
print(f"varInt: {varInt} varFloat: {varFloat}")

# convert float to int
varFloat = 10.9
varInt = int(varFloat)
print(f"varFloat: {varFloat} varInt: {varInt}")
````

# Python and Types

## Combining different types

* we now know about numbers and strings.
* you cannot combine numbers and strings
* to combine two variables need to be the same type.
* you likely want to convert the number to string representation of the original number and then combine

```
>>> var = 10 + '10'

Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
TypeError: unsupported operand type(s) for +: 'int' and 'str'
>>>
```

* When reading from files or API's you will get numbers that are string types.
* You can get a variables type using the `type()` method

```
var = 10
print(f'var is equal to {var} and its type is {type(var)}')

>>> var is equal to 10 and its type is <class 'int'>
```

## Type Conversion

In order to be able to combine: 10 and '10' they need to be same type.
* you can convert to int using the method: `int()`
* you can convert to string using the method: `str()`

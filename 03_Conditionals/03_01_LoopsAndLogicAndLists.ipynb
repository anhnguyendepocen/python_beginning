## List Methods

* like strings lists come with a bunch of methods
* [list methods](https://docs.python.org/3/tutorial/datastructures.html#more-on-lists)

### commonly used list methods
* **append()**   - add elements to a list
* **extend()**   - combine two lists
* **insert()**   - add new elements to a list in a specific position
* **sort()**     - sorts the list
* **pop()**

### using list methods:

#### Adding an element to a list using `append`
```
foodsThatArentHotDogs = ['pizza', 'burger', 'steak']
foodsThatArentHotDogs.append('granola')
```

#### Using `insert` an element to a list
```
foodsThatArentHotDogs = ['pizza', 'burger', 'steak']
foodsThatArentHotDogs.insert(1, 'oranges')
>>> ['pizza', 'oranges', 'burger', 'steak']
```

#### Extend a list using `extend`
```
foodsThatArentHotDogs = ['pizza', 'burger', 'steak']
vegetables = ['lettuce', 'cauliflower', 'broccoli', 'chicken']
foodsThatArentHotDogs.extend(vegetables)
>>> ['pizza', 'burger', 'steak', 'lettuce', 'cauliflower', 'broccoli', 'chicken']
```

## Copy Lists

<img src="https://st2.depositphotos.com/3276065/10679/i/950/depositphotos_106795048-stock-photo-a-red-stamp-copy-on.jpg" width=100>

You can't say `myList = someOtherList` if you want to copy a list.

Example:
```
myList = ['cat', 'dog', 'fish']
petList = myList
petList.append('snake')
print(f"my list is: {myList}")

>>> my list is: ['cat', 'dog', 'fish', 'snake']
```

Instead if you want a copy of a list you need to:
* use copy method
* use a slice to copy

Example:
```
myList = ['cat', 'dog', 'fish']
newList = myList.copy()
petList.append('snake')
print(f"myList without snakes is: {myList}")

>>> myList without snakes is: ['cat', 'dog', 'fish']
```

## The following resolve to `False`

                  'Brett Hull', "Jesus McDavid",  'Guy Lafeur', 
                  'Mike Bossy', 'Phil Esposito', 'Carey Price',

## Combine code blocks `if` and `for` working together.

* example of putting an if statement into a for loop

```
listOfPlayers =  ['Gordie Howe', 'Alex Ovechkin', 'Jean Beliveau',
                  'Brett Hull', "Jesus McDavid",  'Guy Lafeur', 
                  'Mike Bossy', 'Phil Esposito', 'Carey Price', 
                  'Mats Naslund']
for player in listOfPlayers:
    if player[0].upper() == 'G':  
        print(f"Players with names that start with G: {player}")

>>> Players with names that start with G: Gordie Howe
>>> Players with names that start with G: Guy Lafeur
```

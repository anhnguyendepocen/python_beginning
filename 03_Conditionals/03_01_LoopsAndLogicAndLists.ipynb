## List Methods

* like strings lists come with a bunch of methods
* [list methods](https://docs.python.org/3/tutorial/datastructures.html#more-on-lists)

### commonly used list methods
* **append()**   - add elements to a list
* **extend()**   - combine two lists
* **insert()**   - add new elements to a list in a specific position
* **sort()**     - sorts the list
* **pop()**

### using list methods:

#### Adding an element to a list using `append`
```
foodsThatArentHotDogs = ['pizza', 'burger', 'steak']
foodsThatArentHotDogs.append('granola')

>>> ['pizza', 'burger', 'steak', 'granola']
```

#### Using `insert` an element to a list
```
foodsThatArentHotDogs = ['pizza', 'burger', 'steak']
foodsThatArentHotDogs.insert(1, 'oranges')

>>> ['pizza', 'oranges', 'burger', 'steak']
```

#### Extend a list using `extend`
```
foodsThatArentHotDogs = ['pizza', 'burger', 'steak']
vegetables = ['lettuce', 'cauliflower', 'broccoli', 'chicken']
foodsThatArentHotDogs.extend(vegetables)

>>> ['pizza', 'burger', 'steak', 'lettuce', 'cauliflower', 'broccoli', 'chicken']
```

# range() function

* you can also generate sequential lists of numbers using the range function
* can be used with a for loop to iterate over a sequential list of numbers.

## Range 1 Args
* when range gets a single number arguement it generates a sequential list of number up to but not including that number

### Example:
```
numList = range(4)
print(list(numList))

>>> [0, 1, 2, 3]
```

## Range 2 Args
* generates a list of numbers
* list starts at first arg
* goes up to but not including the last arg

### Example
```
numList = list(range(3,6))
print(numList)

>>> [3, 4, 5]
```

## Range 3 Args
* generates a list of numbers
* first arg is the starting number
* second arg is the concluding number
* third arg is the interval

### Example
```
numList = list(range(4, 15, 3))
print(numList)

>>> [4, 7, 10, 13]
```

## Function: Recieving and Returning data.
* functions receive data through arguements
* funcions use `return` statement to return data

### Example

````python
# a function that recieves on arguement
def functionWithArgs(argNumberOne):
    print(f"argNumberOne is {argNumberOne}")

# functions can also recieve multiple arguements
def functionWithMultiArgs(argOne, argTwo, argThree, ArgArg):
    print(f"argOne is: {argOne}")
    print(f"ArgArg is: {ArgArg}")

# functions can return data also
def addUpTheArgs(arg1, arg2):
    newData = arg1 + arg2
    return newData

# returning multiple values
def returnMultipleValues(arg1, arg2):
    newData = arg1 + arg2
    # to return multiple values put the data into list
    returnList = [arg1, arg2, newData]
    return returnList

# call the function that takes one arguement
functionWithArgs('arg sent to function')

# caling a funtion that recieves multiple args, the function 
# "functionWithMultiArgs" is expecting 4 args.  Any less or any more
# will generate an error.
# in the call below:
#    'one' goes into the arg: argOne
#    'two' goes into the arg: argTwo
#    'three' goes into the arg: argThree
#    'four' goes into the arg: ArgArg
functionWithMultiArgs("one", 'two', 'three', 'four')

# calling the add up function but instead of sending the args 
# based on positions we are sending using the names.
# this way the order doesn't matter
# the returned value from the function goes into "joinedValue"
joinedValue = addUpTheArgs(arg2='Is The Greatest', arg1='Guy Lafleur')
print(f"joinedValue is: {joinedValue}")

joinedValuesList = returnMultipleValues(arg2='Is The Greatest', arg1='Guy Lafleur')
print(f"joinedValuesList: {joinedValuesList}")
````

## Function Scoping: Declare / Force global scope

* in the example below we are going so specify global scope in a function for a variable
* when we do this, references to that variable will refer to the global version

### Example:

````python
numVar = 1

if numVar:
    # Scoping on effects variables defined in functions.
    # the var: 'anotherVar' is also global
    anotherVar = 2

def whatsYourFunction(varA, varB):
    # by saying 'global' below we are saying all references to the variable
    # anotherVar should refer to the global version
    global anotherVar
    print(f"    inside function before operation: {anotherVar}")
    anotherVar = varA + varB
    print(f"    inside function after operation: {anotherVar}")

print(f"anotherVar before function: {anotherVar}")
whatsYourFunction(3, 4)
print(f"anotherVar after function: {anotherVar}")
````

# if __name__ == '__main__':

* you will likely come accross this code block
* it allows your code to be used as a library or a tool
* if the module is imported and used elsewhere __name__ will NOT be __main__
* if the module itself is executed, then __name__ will be __main__

````python

def addUptheNums(inList):
    """
    :param inList: a list of numbers.  The list must only contain 
                   numbers
    :return: the sum of all the number in the input list
    """
    sum = 0
    for num in inList:
        sum = sum + num
    return sum

if __name__ == '__main__':
    # these lines call the actual function without 
    # these lines we only have a function definition
    newSum = addUptheNums([34, 23, 22, 7889, 23,1, 3, 877])
    print(f"the sum is {newSum}")
    
````

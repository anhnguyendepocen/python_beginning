## the `response` object contains all the information about the request
* [response object documentation](https://requests.readthedocs.io/en/master/api/#requests.Response)
* [response.status_code](https://requests.readthedocs.io/en/master/api/#requests.Response.status_code): The status code associated with your request.  [more on status codes](https://www.restapitutorial.com/httpstatuscodes.html)
* [response.text](https://requests.readthedocs.io/en/master/api/#requests.Response.text): response text
* [response.json()](https://requests.readthedocs.io/en/master/api/#requests.Response.json): the returned json response converted to python data struct.


### Request Summary:
* **endpoint:** https://api.github.com/users/franTarkenton/gists
* **method**: *GET*
* **params:** *None*
* **headers:** *Accept: application/json*
* **body**: *None*

**Note** The header provided is not required, just adding it to demo how you can add headers.

````python
import requests
import urllib
import json

# The root url for api calls. 
# for url joining... make sure ends with a '/'
rootUrl = 'https://api.github.com/'

# specific end point.
# if you have a github account you can put your name in here..
user = 'franTarkenton'
endpoint = f'users/{user}/gists'
# glue the root url and end point together... (make sure root url has a trailing / or this won't work as expected)
dstUrl = urllib.parse.urljoin(rootUrl, endpoint)
print(f'dstUrl: {dstUrl}')

headers = {'Accept': 'application/json'}

# making a get request
resp = requests.get(dstUrl, headers=headers)
# print the status_code
print(f'status code: {resp.status_code}')

# get the actual data from the request object
jsonData = resp.json()

# dump the data to a file
# a lot of data so dump to file for inspection
outFile = f'github_{user}_gists.json'
with open(outFile, 'w') as fh:
    json.dump(jsonData, fh, sort_keys=True, indent=4)
print(f"num gists returned: {len(jsonData)}")
````

## Simple GET request with parameters

### Request Summary:
* **endpoint:** https://catalog.data.gov.bc.ca/api/3/action/package_list
* **method**: GET
* **params:** limit=11, offset=1
* **headers:** None
* **body**: None



````python
import requests
import urllib
import json
import pprint

# The root url for api calls.
rootUrl = 'https://catalog.data.gov.bc.ca/api/3/action/'
# specific end point.
endpoint = 'package_list'
# glue the root url and end point together... (make sure root url has a trailing / or this won't work as expected)
dstUrl = urllib.parse.urljoin(rootUrl, endpoint)
print(f'dstUrl: {dstUrl}')

# specify the parameters in a dict
params = {'limit': 11,
          'offset': 0}
          
# making a get request, and sending the params
resp = requests.get(dstUrl, params=params)
# print the status_code
print(f'status code: {resp.status_code}')

# get the actual data from the request object
jsonData = resp.json()

# only have 11 records so can print to screen
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(jsonData)
````
